```
imdb = pd.read_csv('imdb-reviews-pt-br.csv',encoding='utf-8')
```

```
imdb_pt = imdb[['text_pt','sentiment']]
imdb_pt.head()
```

```
imdb_pt.sentiment.unique()
```

```
imdb_pt.sentiment.replace( {'neg': 0,'pos': 1}, inplace=True )
```

```
imdb_pt
```

```
imdb_reviews = imdb_pt.text_pt.values      #x
imdb_sentiment =imdb_pt.sentiment.values   #y
```

```
x_train,x_test,y_train,y_test = train_test_split(imdb_reviews,imdb_sentiment,test_size=0.2,random_state=42)
```

```
y_test
```

```
x_test
```

```
#Bag of words com stopwords

vectorize = CountVectorizer(analyzer = 'word',stop_words={'portuguese'})
vectorize.fit(imdb_reviews)
```

```
bow_x_train = vectorize.transform(x_train)
bow_x_test  = vectorize.transform(x_test)
```

```
bow_x_train
```

##Classificador

```
model = MultinomialNB()
model.fit(bow_x_train,y_train)
```

##Test

```
txt_teste = ['O melhor filme do ano. Muito bom mesmo, atores de alto nível.','Filme Horroroso,coisa ruim!']
meu_teste = vectorize.transform(txt_teste).toarray()
saida = model.predict(meu_teste)
print(saida)
```
